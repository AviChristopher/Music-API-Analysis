```python
import pandas as pd
import matplotlib.pyplot as plt



movies =pd.read_csv('updated_movies.csv')
movies.head()

movies["Runtime"]=pd.to_numeric(movies["Runtime"])
                           
```

```python
movies.info()
```

```python
#Cleaned the data

updated_movies = movies.copy()
updated_movies.loc[0, 'Genre'] = 'Biography,Comedy,Drama'
updated_movies.loc[0, 'Movie Title'] = 'Green Book'
updated_movies.loc[0, 'Runtime'] = '130'
```

```python
updated_movies.head(15)
```

```python
new_data = {'Movie Title': 'Parasite', 'Runtime': 132, 'Genre': 'Drama, Thriller', 'Award Wins': 4, 'Award Nominations': 306 , 'Box Office':53369749}

# Add the new row to the DataFrame
updated_movies =updated_movies.append(new_data, ignore_index=True)
```

```python
movies=updated_movies
movies.head()
movies.to_csv('updated_movies.csv', index=False)
```

# 1. What was the runtime for each movie?

#### EXPLANATION

##### As shown from the data below, the mean, the median, the lowest,highest and standard deviation of Runtime was calculated.

##### The average/mean or median runtime is around 121 minutes. 

##### <ins> 'The Artist'</ins> has the lowest runtime with 100 minutes and  <ins>'12 years a Slave'</ins> had the longest Runtime with 134 minutes.

####     <div align="center">        <ins> Data below showing the Title of the movie and its Runtime </ins>




```python
runtime_movies=movies.loc[:,['Movie Title','Runtime']]
print(runtime_movies)
```

#### <div align="center"> Calculating the mean,median,minimum, maximum and Standard deviation of the Runtime column from the dataset.

```python
mean_runtime = movies['Runtime'].mean()
median_runtime = movies['Runtime'].median()
min_runtime = movies['Runtime'].min()
max_runtime = movies['Runtime'].max()
std_dev_runtime = movies['Runtime'].std()

print(f"Mean Runtime: {mean_runtime} minutes")
print(f"Median Runtime: {median_runtime} minutes")
print(f"Minimum Runtime: {min_runtime} minutes")
print(f"Maximum Runtime: {max_runtime} minutes")
print(f"Standard Deviation of Runtime: {std_dev_runtime}")
```

###     <div align="center">        <ins> The Line graph below shows the runtime for each movie </ins>

```python
# Plotting a line graph
plt.figure(figsize=(12, 6))
plt.plot(movies['Movie Title'], movies['Runtime'], marker='o', color='blue', linestyle='-')
plt.title('Runtime of Movies')
plt.xlabel('Movie Title')
plt.ylabel('Runtime (minutes)')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability

plt.show()
```

###    <div align="center">        <ins> The histogram below shows the distribution of Movie Runtimes </ins>

##### The majority of movies are between 115 and 120 minutes.

```python
import matplotlib.pyplot as plt

plt.hist(movies['Runtime'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Runtimes')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

# 3. Is there any relationship between the length of the movie (runtime) and the number of awards a movie won?


#### Answer
##### The scatter plot below shows a clear negative correlation between Runtime and the number of awards a movie has won.

```python
import matplotlib.pyplot as plt

plt.scatter(movies['Runtime'], movies['Award Wins'], color='blue', marker='o')
plt.title('Scatter Plot of Runtime vs Award Wins')
plt.xlabel('Runtime')
plt.ylabel('Award Wins')
plt.show()
```
